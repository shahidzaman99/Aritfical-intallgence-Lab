```python
# DFS Program for given graph (0,1,2,3,4)

graph = {
    0: [1, 2, 3],
    1: [0],
    2: [0, 3, 4],
    3: [0, 2],
    4: [2]
}

visited = set()

def dfs(node):
    if node not in visited:
        print("Visited:", node)
        visited.add(node)

        print("Adjacent nodes of", node, ":", graph[node])

        for neighbor in graph[node]:
            dfs(neighbor)


print("DFS Traversal Order:\n")
dfs(0)   # Start from node 0
```

```python
# BFS Program for given graph (0,1,2,3,4)

from collections import deque

graph = {
    0: [1, 2, 3],
    1: [0],
    2: [0, 3, 4],
    3: [0, 2],
    4: [2]
}

visited = set()

def bfs(start):
    queue = deque([start])
    visited.add(start)

    print("BFS Traversal Order:\n")

    while queue:
        node = queue.popleft()
        print("Visited:", node)
        print("Adjacent nodes of", node, ":", graph[node])

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)


# Start BFS from node 0
bfs(0)
```
