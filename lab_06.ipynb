```python
import heapq

def a_star_graph_search(graph, heuristics, start, goal):
    # Priority queue stores (f_score, current_node, path_taken)
    # f = g (actual distance) + h (heuristic estimate)
    priority_queue = [(heuristics[start], start, [start], 0)]
    visited = set()

    while priority_queue:
        f, current, path, g = heapq.heappop(priority_queue)

        if current == goal:
            return path, g

        if current not in visited:
            visited.add(current)
            for neighbor, weight in graph[current].items():
                if neighbor not in visited:
                    new_g = g + weight
                    new_f = new_g + heuristics[neighbor]
                    new_path = path + [neighbor]
                    heapq.heappush(priority_queue, (new_f, neighbor, new_path, new_g))

    return None, float('inf')

# Define Map: Cities and the actual distance (g) between them
graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'C': 2, 'D': 5},
    'C': {'A': 4, 'B': 2, 'D': 1},
    'D': {'B': 5, 'C': 1}
}

# Define Heuristics (h): Estimated straight-line distance to Goal 'D'
heuristics = {
    'A': 7,
    'B': 6,
    'C': 1,
    'D': 0
}

# Run Search
path, total_cost = a_star_graph_search(graph, heuristics, 'A', 'D')
print(f"Shortest Path from A to D: {path}")
print(f"Total Movement Cost (g): {total_cost}")
```

```python
# =========================
# TASK 1: 8-Puzzle A* (Misplaced Tiles)
# =========================

# Puzzle setup using append
puzzle = []
puzzle.append(2)
puzzle.append(8)
puzzle.append(3)
puzzle.append(1)
puzzle.append(6)
puzzle.append(4)
puzzle.append(7)
puzzle.append(0)
puzzle.append(5)

goal = []
goal.append(1)
goal.append(2)
goal.append(3)
goal.append(8)
goal.append(0)
goal.append(4)
goal.append(7)
goal.append(6)
goal.append(5)

# Heuristic: number of misplaced tiles
def misplaced_tiles(state, goal):
    count = 0
    for i in range(9):
        if state[i] != 0 and state[i] != goal[i]:
            count += 1
    return count

# Possible moves for blank
def moves(state):
    index = state.index(0)
    result = []
    directions = [-3, 3, -1, 1]  # up, down, left, right
    for d in directions:
        new = index + d
        if new < 0 or new > 8:
            continue
        if index % 3 == 0 and d == -1:
            continue
        if index % 3 == 2 and d == 1:
            continue
        new_state = state.copy()
        new_state[index], new_state[new] = new_state[new], new_state[index]
        result.append(new_state)
    return result

# A* Algorithm
def evaluate_misplaced(start, goal):
    open_list = []
    open_list.append((0, start, []))  # f(n), state, path
    visited = []

    while open_list:
        open_list.sort()
        f, state, path = open_list.pop(0)
        if state == goal:
            return path + [state]
        if state in visited:
            continue
        visited.append(state)
        g = len(path) + 1
        for m in moves(state):
            f_val = g + misplaced_tiles(m, goal)
            open_list.append((f_val, m, path + [state]))

# Solve puzzle
solution = evaluate_misplaced(puzzle, goal)

# Print solution
print("===== TASK 1: 8-Puzzle Solution =====\n")
step_num = 0
for s in solution:
    print(f"Step {step_num}:")
    print(s[0:3])
    print(s[3:6])
    print(s[6:9])
    print()
    step_num += 1
```

```python
# =========================
# TASK 2: Graph Shortest Path A*
# =========================

# Graph: nodes with distances
graph = {
    'A': {'B': 1, 'C': 3},
    'B': {'A': 1, 'D': 1, 'E': 4},
    'C': {'A': 3, 'F': 2},
    'D': {'B': 1},
    'E': {'B': 4, 'F': 1},
    'F': {'C': 2, 'E': 1}
}

# Heuristics: estimated distance to goal F
heuristics = {
    'A': 6,
    'B': 4,
    'C': 2,
    'D': 5,
    'E': 1,
    'F': 0
}

# A* Algorithm for graph
def a_star_graph(start, goal, graph, heuristics):
    open_list = []
    open_list.append((heuristics[start], start, [], 0))  # f, node, path, g
    visited = []

    while open_list:
        open_list.sort()
        f, node, path, g = open_list.pop(0)
        if node == goal:
            return path + [node], g
        if node in visited:
            continue
        visited.append(node)
        for neighbor, cost in graph[node].items():
            new_g = g + cost
            new_f = new_g + heuristics[neighbor]
            open_list.append((new_f, neighbor, path + [node], new_g))

# Solve shortest path
path, total_cost = a_star_graph('A', 'F', graph, heuristics)

print("===== TASK 2: Graph Shortest Path =====\n")
print("Shortest Path:", path)
print("Total Cost:", total_cost)
```
