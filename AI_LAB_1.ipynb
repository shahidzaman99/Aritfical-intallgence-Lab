```python
def is_prime(number):
  if number <= 1:
    return False
  for i in range (2, int (number ** 0.5) + 1):
    if number % i == 0:
      return False
  return True

num = int(input("Enter a number: "))
if is_prime(num):
  print(f"The number {num} is prime.")
else:
  print(f"The number {num} is not prime.")
```

```python
count = 0
for i in range (2, 101):
  if is_prime(i):
    count += 1
    print (count,"\t",i)
    if count >= 20:
      break
```
